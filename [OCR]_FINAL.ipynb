```python
import os
import cv2
import easyocr
import pytesseract
import re
import pandas as pd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt

# Initialize EasyOCR
easyocr_reader = easyocr.Reader(['en'])

# Define boundary points for the agreement line (협정선 동측 한계선)
boundary_points = [
    (125 + 25 / 60, 32 + 11 / 60),
    (124 + 8 / 60, 33 + 20 / 60),
    (124 + 0 / 60 + 30 / 3600, 34),
    (124 + 7 / 60 + 30 / 3600, 35),
    (124 + 30 / 60, 35 + 30 / 60),
    (124 + 30 / 60, 36 + 45 / 60),
    (124 + 20 / 60, 37),
    (126 + 45 / 60, 32 + 11 / 60)
]
boundary_polygon = Polygon(boundary_points)

# Function to classify vessel position
def classify_vessel_position(longitude, latitude):
    vessel_point = Point(longitude, latitude)
    if boundary_polygon.contains(vessel_point):
        return "허가"  # Inside the boundary
    else:
        return "불법"  # Outside the boundary

# Function to perform OCR and extract coordinates from the image
def extract_coordinates_from_image(image_path):
    # Load and preprocess the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to load image {image_path}")
        return "Longitude not found", "Latitude not found", None, None

    # Step 1: Crop the image to focus on coordinates
    height, width = image.shape[:2]
    crop_height_ratio = 0.04
    cropped_image = image[int(height * (1 - crop_height_ratio)):height, 0:int(width * 0.5)]


# Define cropping ratios
crop_height_ratio = 0.04  # Proportion to crop from the top
crop_left_ratio = 0.05  # Proportion to crop from the left
crop_right_ratio = 0.05  # Proportion to crop from the right

# Calculate cropping dimensions
crop_top = int(height * (1 - crop_height_ratio))
crop_bottom = height
crop_left = int(width * crop_left_ratio)
crop_right = int(width * (1 - crop_right_ratio))

# Perform cropping
cropped_image = image[crop_top:crop_bottom, crop_left:crop_right]

     # Display the cropped image
    plt.figure(figsize=(10, 5))
    plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    plt.title(f"Cropped Image: {os.path.basename(image_path)}")
    plt.axis("off")
    plt.show()

    # Step 2: Convert to grayscale and resize for better OCR
    gray = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
    target_width = 3000
    aspect_ratio = target_width / gray.shape[1]
    resized_image = cv2.resize(gray, (target_width, int(gray.shape[0] * aspect_ratio)), interpolation=cv2.INTER_LINEAR)

    # Perform OCR with EasyOCR
    easyocr_result = easyocr_reader.readtext(resized_image, detail=0)
    easyocr_text = " ".join(easyocr_result)

    # Perform OCR with Tesseract
    tesseract_text = pytesseract.image_to_string(resized_image, config="--psm 6")

    # Combine OCR outputs
    combined_text = easyocr_text if len(easyocr_text) > len(tesseract_text) else tesseract_text
    print(f"OCR Output for {os.path.basename(image_path)}:")
    print(combined_text)

    # Clean the OCR output text
    cleaned_text = combined_text.replace(",", ".")
    cleaned_text = re.sub(r"[^0-9NSE°.\s']", "", cleaned_text)
    print(f"Cleaned OCR Output for {os.path.basename(image_path)}:")
    print(cleaned_text)

    # Extract coordinates using regex
    lat_pattern = r"N\s*(\d{1,2})°\s*(\d{1,2}\.\d+)"
    lon_pattern = r"E\s*(\d{1,3})°\s*(\d{1,2}\.\d+)"

    lat_match = re.search(lat_pattern, cleaned_text)
    lon_match = re.search(lon_pattern, cleaned_text)

    if lat_match and lon_match:
        # Convert to decimal format
        latitude = int(lat_match.group(1)) + float(lat_match.group(2)) / 60
        longitude = int(lon_match.group(1)) + float(lon_match.group(2)) / 60
        return f"N {lat_match.group(1)}° {lat_match.group(2)}'", f"E {lon_match.group(1)}° {lon_match.group(2)}'", longitude, latitude
    else:
        print(f"Coordinates not found in the OCR output for {os.path.basename(image_path)}.")
        return "Longitude not found", "Latitude not found", None, None

# Batch processing function
def batch_process_images(image_folder_path):
    results = []

    # Iterate through all image files in the folder
    for file_name in os.listdir(image_folder_path):
        if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(image_folder_path, file_name)
            print(f"Processing: {file_name}")

            # Step 1: Perform OCR to extract coordinates
            latitude_dms, longitude_dms, latitude_decimal, longitude_decimal = extract_coordinates_from_image(image_path)

            # Step 2: Classify the coordinates
            if longitude_decimal is not None and latitude_decimal is not None:
                classification = classify_vessel_position(latitude_decimal, longitude_decimal)
                location = f"{latitude_dms} / {longitude_dms}"  # Ensure N comes first
            else:
                # Default to LAT N0.0.0 LON E0.0.0 and classify as 불법
                classification = "불법"
                location = "LAT N0.0.0 / LON E0.0.0"

            # Append results for each image
            results.append({
                "파일명": file_name,
                "위치": location,
                "분류": classification
            })

    # Convert results to a pandas DataFrame for display
    df = pd.DataFrame(results)
    print("\n--- Results ---\n")
    print(df)

    # Example usage
image_folder_path = "/content/drive/MyDrive/final_data_nameChanged/data_nameChanged/testforOCR"
batch_process_images(image_folder_path)
```
